<H2>.APPLY</H2>

```python
df = pd.read_csv('train.csv')

def contract(x):
    if x == 'male':
        return 'M'
    
    if x == 'female':
        return 'F'
    
```

```python
df["Sex"].apply(lambda x:contract(x))
```

```python
df["Sex"] = df["Sex"].apply(lambda x:contract(x))
```
