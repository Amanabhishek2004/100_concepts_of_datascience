```python
encodings_list = [
    'utf-8',
    'latin-1',
    'utf-16',
    'utf-32',
    'ascii',
    'cp1252',
    'gbk',
    'big5',
    'shift-jis',
]
```

```python
for encoding in encodings_list:
   
   try:
       pd.read_csv("spam.csv" , encoding = encoding)
       print(f"worked with {encoding}")
   except UnicodeDecodeError:  
         print(f" not the correct form of {encoding}")
```
