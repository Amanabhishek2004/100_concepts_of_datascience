<h1>.REPLACE</h1>

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1]
})
```

```python
df
```

```python
#  replacing all rows value
df_replaced = df.replace(1, 100)
df_replaced
```

```python
# multiple values
df_replaced = df.replace([1, 4], [100, 400])
df_replaced
```

```python
# specific cloumn
df_replaced = df.replace({'A': 1, 'B': 4}, 100)
df_replaced
```

```python
# multiple values in specific column
df.replace({"A":{1:100} ,"B":{4:200}})
```
